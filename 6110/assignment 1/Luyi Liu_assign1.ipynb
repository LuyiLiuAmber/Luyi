```python
#Replace Missing Values
df['Product'] = df['Product'].fillna('Other')
```

```python
df['Region'] = df['Region'].fillna("Other")
```

```python
df['Price'] = df['Price'].fillna(df['Price'].mean())
```

```python
df['Units Sold'] = df['Units Sold'].fillna(df['Units Sold'].mean())
```
