```python
df_grouped = df.groupby("brand")
df_grouped.groups
```

```python
df_grouped.get_group('apple')
```

```python
df_grouped.get_group('samsung')
```

H0 = Apple's price is lower than Samsung
H1 = Apple's price is higher than Samsung

```python
apple = df[df['brand'] == 'apple']['price']
samsung = df[df['brand'] == 'samsung']['price']
```

```python
t_stat, p_val = stats.ttest_ind(apple, samsung)
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")

```

as P<0.05 reject H0, so samsung price more than apple
